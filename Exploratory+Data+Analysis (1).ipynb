# Exploratory Data Analysis

```python
#importing all the essential packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
#Reading the CSV
data = pd.read_csv('C:/USF_BAIS/DSP/Twitter/output1.csv',encoding = 'ISO-8859-1')
```

```python
#Checking top rows
data.head()
```

```python
#Classifying Bots and NoBots into Separate dataframe
bots = data[data.Bot_or_NoBot==0]
nonbots = data[data.Bot_or_NoBot==1]
```

```python
#Checking bot enteries
bots
```

```python
#Scatter plot of Friends count vs Followers count of both bot and non bot
bots.user_friends_count/bots.user_followers_count

plt.figure(figsize=(10,5))
plt.subplot(2,1,1)
plt.title('Bots Friends vs Followers')
sb.regplot(bots.user_friends_count, bots.user_followers_count, color='red', label='Bots')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.tight_layout()

plt.subplot(2,1,2)
plt.title('NonBots Friends vs Followers')
sb.regplot(nonbots.user_friends_count, nonbots.user_followers_count, color='blue', label='NonBots')
plt.xlim(0, 100)
plt.ylim(0, 100)

plt.tight_layout()
plt.show()

#below line taken from https://github.com/jubins/MachineLearning-Detecting-Twitter-Bots/blob/master/FinalProjectAndCode/IPython%20NoteBooks/BotDetection.ipynb
```

```python
bots['friends_by_followers'] = bots.user_friends_count/bots.user_followers_count
bots[bots.friends_by_followers<1].shape

nonbots['friends_by_followers'] = nonbots.user_friends_count/nonbots.user_followers_count
nonbots[nonbots.friends_by_followers<1].shape
```

```python
#listed count of bots and non bot
plt.figure(figsize=(10,5))
plt.plot(bots.user_listed_count, color='red', label='Bots')
plt.plot(nonbots.user_listed_count, color='blue', label='NonBots')
plt.legend(loc='upper left')
plt.ylim(1000,2000)
print(bots[(bots.user_listed_count<5)].shape)


#below line taken from https://github.com/jubins/MachineLearning-Detecting-Twitter-Bots/blob/master/FinalProjectAndCode/IPython%20NoteBooks/BotDetection.ipynb
```

```python
bots_listed_count_df = bots[bots.user_listed_count<16000]
nonbots_listed_count_df = nonbots[nonbots.user_listed_count<16000]

bots_verified_df = bots_listed_count_df[bots_listed_count_df.user_verified==False]
bots_screenname_has_bot_df_ = bots_verified_df[(bots_verified_df.user_screen_name.str.contains("bot", case=False)==True)].shape
```

```python
#friends and followers of bot and non bot
plt.figure(figsize=(12,7))

plt.subplot(2,1,1)
plt.plot(bots_listed_count_df.user_friends_count, color='red', label='Bots Friends')
plt.plot(nonbots_listed_count_df.user_friends_count, color='blue', label='NonBots Friends')
plt.legend(loc='upper left')

plt.subplot(2,1,2)
plt.plot(bots_listed_count_df.user_followers_count, color='red', label='Bots Followers')
plt.plot(nonbots_listed_count_df.user_followers_count, color='blue', label='NonBots Followers')
plt.legend(loc='upper left')

#below line taken from https://github.com/jubins/MachineLearning-Detecting-Twitter-Bots/blob/master/FinalProjectAndCode/IPython%20NoteBooks/BotDetection.ipynb
```

```python

df = pd.concat([bots, nonbots])
df.shape
```

```python
#corellation matrix
df.corr(method='spearman')
```

```python
#shows that followers count and statuses count have tge maximum correlation also listed count and followers count have the maximum corellation
plt.figure(figsize=(8,4))
sb.heatmap(df.corr(method='spearman'), cmap='coolwarm', annot=True)
plt.tight_layout()
plt.show()
```
